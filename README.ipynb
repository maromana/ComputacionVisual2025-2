## 3. Iluminación aplicada
  ### Esquema key–fill–rim

**Luz principal (Key Light)**  
- Tipo: `DirectionalLight`  
- Posición: (X=5, Y=10, Z=7,5)  
- Color: tono cálido (anaranjado suave)  
- Intensidad: ~1.0  

**Luz de relleno (Fill Light)**  
- Tipo: `DirectionalLight`  
- Posición: (X=-5, Y=5, Z=7,5)  
- Color: más frío (azulado frio)  
- Intensidad: ~0.5

**Luz de contorno (Rim Light)**  
- Tipo: `DirectionalLight`
- Posición: (X=5, Y=10, Z=4.4)
- Color: blanco neutro
- Intensidad: ~0.8  
- Función: resaltar bordes y dar profundidad  

**Luz ambiental (Ambient Light)**  
- Tipo: `AmbientLight`  
- Posición: (X=0, Y=0, Z=0)
- Intensidad: ~0.2

## 4. Materiales y texturas (PBR)
- Parámetros principales:  
  - `roughness`: se modifica de acuerdo a la zona, para representar la dureza de un material.  
  - Normal map en el objeto de la lampara y en el edificio para dar textura como la madera o grietas en el edificio en ruinas.
- Justificación: Expresar el realismo de los materiales, como la madera vieja, una pared agrietada, entre otros.

---
Se adjunta a continuación una tabla con las texturas manejadas

```python
import base64
from IPython.display import HTML

# Función para convertir imagen a base64
def img_to_base64(path):
    with open(path, "rb") as f:
        data = f.read()
    return base64.b64encode(data).decode()

# Datos de ejemplo (ajusta las rutas a tus imágenes en /content/)
texturas = ["Edificio",  "Lampara", "Madera", "Hornet Cabeza", "Hornet Arma", "Hornet Cuerpo", "Hornet Ropa"]
imagenes = ["/content/texturaedifcio.png",   # cambia esta ruta por tu archivo subido
        "/content/textura2lamp.png",
        "/content/textura1lamp.png",
        "/content/hornetcabeza.png",
        "/content/hornetaguijon.png",
        "/content/hornetcuerpo.png",
        "/content/hornetvestido.png"]  # cambia por tus archivos
lugares = ["Edificio Pared", "Lampara", "Poste", "Cabeza Personaje", "Aguja que sujeta personaje", "Cuerpo debajo de la capa roja", "Capa roja del personaje"]

# Construcción de la tabla
html = "<table border='1' style='border-collapse:collapse; text-align:center;'>"
html += "<tr><th>Textura</th><th>Imagen</th><th>Lugar</th></tr>"

for t, img, l in zip(texturas, imagenes, lugares):
    try:
        img64 = img_to_base64(img)
        img_tag = f"<img src='data:image/jpeg;base64,{img64}' width='100'>"
    except:
        img_tag = "(No encontrada)"
    html += f"<tr><td>{t}</td><td>{img_tag}</td><td>{l}</td></tr>"

html += "</table>"

# Mostrar
HTML(html)
```

## 8. Modelo de color
- Paleta definida en formato HEX a continuación m
- Justificación perceptual: se basa en las paletas básicas del videojuego y del personaje.  

---

```python
import matplotlib.pyplot as plt

# Colores principales extraídos (HEX)
colors_hex = ["#C02E3C", "#D8D2C7", "#2B2B2B", "#5C3B25", "#7A7A7A"]
labels = ["Rojo capa", "Blanco hueso", "Gris oscuro", "Marrón madera", "Gris cemento"]

# Crear figura
fig, ax = plt.subplots(figsize=(8, 2))

# Dibujar los rectángulos de color
for i, (hex_color, label) in enumerate(zip(colors_hex, labels)):
    ax.add_patch(plt.Rectangle((i, 0), 1, 1, color=hex_color))
    ax.text(i + 0.5, -0.2, label, ha='center', va='top', fontsize=10)
    ax.text(i + 0.5, 0.5, hex_color, ha='center', va='center', fontsize=9, color="white" if i in [0,2,3] else "black")

# Ajustar vista
ax.set_xlim(0, len(colors_hex))
ax.set_ylim(0, 1)
ax.axis("off")

plt.show()
```
